```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel("grid.xlsx")

# Create weighted graph (Failure Rate as edge weight)
G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['node1'], row['node2'], weight=row['Failure rate'])

# --- Kruskal's Algorithm ---
mst_kruskal = nx.minimum_spanning_tree(G, algorithm='kruskal')
total_failure_kruskal = sum(w['weight'] for _, _, w in mst_kruskal.edges(data=True))

# --- Prim's Algorithm ---
mst_prim = nx.minimum_spanning_tree(G, algorithm='prim')
total_failure_prim = sum(w['weight'] for _, _, w in mst_prim.edges(data=True))

# --- Comparison ---
print("🔹 Kruskal's MST:")
for u, v, w in mst_kruskal.edges(data=True):
    print(f"{u} -- {v},\t")
print("Total Failure Rate (Kruskal):", round(total_failure_kruskal, 4))

print("\n🔹 Prim's MST:")
for u, v, w in mst_prim.edges(data=True):
    print(f"{u} -- {v},\t")
print("Total Failure Rate (Prim):", round(total_failure_prim, 4))

# Check if both MSTs are identical
same_edges = set(mst_kruskal.edges()) == set(mst_prim.edges())
print("\n Are both MSTs identical?", "Yes" if same_edges else "No")

# --- Visualization ---
pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(14, 6))

# Kruskal MST
plt.subplot(1, 2, 1)
nx.draw(mst_kruskal, with_labels=True, node_color='red',node_size=300,font_size=8.5)
plt.title("Kruskal's MST (Failure Rate)")

# Prim MST
plt.subplot(1, 2, 2)
nx.draw(mst_prim, with_labels=True, node_color='green', node_size=300,font_size=8.5)
plt.title("Prim's MST (Failure Rate)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel("grid.xlsx")

# Create graph using line length as edge weight
G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['node1'], row['node2'], weight=row['Length (km)'])

# --- Compute connectivity metrics ---
edge_conn = nx.edge_connectivity(G)
vertex_conn = nx.node_connectivity(G)

print(" Edge Connectivity of the Grid (based on Length):", edge_conn)
print("Vertex Connectivity of the Grid (based on Length):", vertex_conn)

# --- Identify critical edges and nodes ---
critical_edges = list(nx.minimum_edge_cut(G))
critical_nodes = list(nx.minimum_node_cut(G))

print("\n Critical Transmission Lines (edges whose removal disconnects the grid):")
for edge in critical_edges:
    length = G[edge[0]][edge[1]]['weight']
    print(f"Line {edge[0]} -- {edge[1]}  (Length: {length} km)")

print("\n Critical Substations (nodes whose failure disconnects the grid):")
print(critical_nodes)

# --- Visualization ---
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10, 7))

# Draw the base graph
nx.draw(G, pos, with_labels=True, node_color='purple', node_size=100,font_size=8.5)

# Highlight critical parts
nx.draw_networkx_edges(G, pos, edgelist=critical_edges, edge_color='red', width=10, label='Critical Lines')
nx.draw_networkx_nodes(G, pos, nodelist=critical_nodes, node_color='orange',node_size=100, label='Critical Substations')

plt.title("Grid Connectivity Analysis Based on Line Length", fontsize=13, fontweight='bold')
plt.legend()
plt.show()
```

```python
import pandas as pd
import networkx as nx

# Load dataset
df = pd.read_excel("grid.xlsx")

# Create weighted graph (Length (km) as weight)
G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['node1'], row['node2'], weight=row['Length (km)'])

# --- Ask user or define source and target nodes ---
source_node_str = input("Enter source node: ").strip()
target_node_str = input("Enter target node: ").strip()

# Convert input strings to integers
try:
    source_node = int(source_node_str)
    target_node = int(target_node_str)
except ValueError:
    print("Invalid input. Please enter integer values for nodes.")
    exit() # Exit or handle the error as appropriate

# --- Run Dijkstra’s Algorithm ---
try:
    shortest_path = nx.dijkstra_path(G, source=source_node, target=target_node, weight='weight')
    shortest_distance = nx.dijkstra_path_length(G, source=source_node, target=target_node, weight='weight')

    print(f"\n🔹 Shortest path from {source_node} to {target_node}: {shortest_path}")
    print(f"📏 Total Length (km): {shortest_distance}")

    # --- Optional: show all shortest paths from source to all other nodes ---
    print("\n📊 Shortest distances from source to all nodes:")
    for node, distance in nx.single_source_dijkstra_path_length(G, source_node, weight='weight').items():
        print(f"{source_node} → {node} = {distance:.3f} km")

except nx.NodeNotFound:
    print(f"Error: One or both nodes ({source_node}, {target_node}) not found in the graph.")
except nx.NetworkXNoPath:
    print(f"No path exists between nodes {source_node} and {target_node}.")


import matplotlib.pyplot as plt

# Compute shortest path and its edges
path_edges = list(zip(shortest_path[:-1], shortest_path[1:]))

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10, 7))

# Draw all edges and nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700)

# Highlight shortest path
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3, label='Shortest Path')
nx.draw_networkx_nodes(G, pos, nodelist=shortest_path, node_color='orange')

# Add edge labels for length
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

plt.title(f"Shortest Path from {source_node} to {target_node}", fontsize=13, fontweight='bold')
plt.legend()
plt.show()

```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# --- Step 1: Load dataset ---
df = pd.read_excel("grid.xlsx")

# --- Step 2: Create graph ---
G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['node1'], row['node2'], weight=row['Failure rate'])

# --- Step 3: Compute betweenness centrality (weighted by failure rate) ---
# Note: In reliability analysis, lower failure rate = better reliability
# so we can use the inverse of failure rate as distance if needed.
bet_nodes = nx.betweenness_centrality(G, weight='Failure rate', normalized=True)
bet_edges = nx.edge_betweenness_centrality(G, weight='Failure rate', normalized=True)

# --- Step 4: Display results ---
print("\n🔹 Node Betweenness Centrality (Critical Substations):")
for node, value in sorted(bet_nodes.items(), key=lambda x: x[1], reverse=True):
    print(f"{node}: {value:.6f}")

print("\n🔹 Edge Betweenness Centrality (Critical Transmission Lines):")
for edge, value in sorted(bet_edges.items(), key=lambda x: x[1], reverse=True):
    print(f"{edge}: {value:.6f}")

# --- Step 5: Visualization ---
plt.figure(figsize=(10, 8))

# Node size proportional to betweenness
node_size = [5000 * bet_nodes[n] + 200 for n in G.nodes()]
edge_width = [5 * bet_edges[e] for e in G.edges()]

pos = nx.spring_layout(G, seed=42)  # for consistent layout

nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color='orange', alpha=0.8)
nx.draw_networkx_edges(G, pos, width=edge_width, edge_color='gray', alpha=0.6)
nx.draw_networkx_labels(G, pos, font_size=9, font_weight='bold')
plt.title("Power Grid Betweenness Centrality\n(Critical Substations and Transmission Lines)", fontsize=14)
plt.axis("off")
plt.show()
```

```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel("grid.xlsx")

# Create graph with failure rate & repair time
G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['node1'], row['node2'],
               weight=row['Failure rate'],
               repair=row['Repair Time (h)'])

# --- Kruskal MST ---
mst_kruskal = nx.minimum_spanning_tree(G, algorithm='kruskal')
total_failure_kruskal = sum(w['weight'] for _, _, w in mst_kruskal.edges(data=True))
avg_repair_kruskal = np.mean([w['repair'] for _, _, w in mst_kruskal.edges(data=True)])

# --- Prim MST ---
mst_prim = nx.minimum_spanning_tree(G, algorithm='prim')
total_failure_prim = sum(w['weight'] for _, _, w in mst_prim.edges(data=True))
avg_repair_prim = np.mean([w['repair'] for _, _, w in mst_prim.edges(data=True)])

# --- Compute Reliability & Availability ---
def compute_reliability_availability(total_failure, avg_repair):
    t = 1  # 1 year
    R = np.exp(-total_failure * t)
    MTBF = 1 / total_failure
    MTTR = avg_repair / 8760  # convert hours to years
    A = MTBF / (MTBF + MTTR)
    return R, A

Rk, Ak = compute_reliability_availability(total_failure_kruskal, avg_repair_kruskal)
Rp, Ap = compute_reliability_availability(total_failure_prim, avg_repair_prim)

# --- Print results ---
print("🔹 Kruskal's MST:")
print(f"Total Failure Rate: {total_failure_kruskal:.5f}")
print(f"Average Repair Time: {avg_repair_kruskal:.2f} h")
print(f"Reliability (1 year): {Rk:.4f}")
print(f"Availability: {Ak*100:.4f}%")

print("\n🔹 Prim's MST:")
print(f"Total Failure Rate: {total_failure_prim:.5f}")
print(f"Average Repair Time: {avg_repair_prim:.2f} h")
print(f"Reliability (1 year): {Rp:.4f}")
print(f"Availability: {Ap*100:.4f}%")

same = set(mst_kruskal.edges()) == set(mst_prim.edges())
print("\n✅ Are both MSTs identical?", "Yes" if same else "No")

# --- Bar chart comparison ---
labels = ['Kruskal', 'Prim']
reliability = [Rk, Rp]
availability = [Ak * 100, Ap * 100]

plt.figure(figsize=(10,5))

# Plot reliability
plt.subplot(1, 2, 1)
plt.bar(labels, reliability, color=['yellow', 'purple'])
plt.title("Reliability Comparison (1 Year)")
plt.ylabel("Reliability (Probability of No Failure)")
plt.ylim(0, 1)

# Plot availability
plt.subplot(1, 2, 2)
plt.bar(labels, availability, color=['blue', 'green'])
plt.title("Availability Comparison")
plt.ylabel("Availability (%)")
plt.ylim(99.5, 100)  # zoom in for small differences

plt.suptitle("Comparison of Kruskal vs Prim (Failure Rate-Based MST)", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
